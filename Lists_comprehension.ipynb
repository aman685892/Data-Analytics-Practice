```python
l1=[30,40,50,60]
l2=[]
for i in l1:
  if i>45:
   l2.append(i)
print(l2)
```

```python
print(l1)
```

```python
l3=[i for i in l1] ## using lists comprehension
print(l3)
```

```python
l3=[i for i in l1 if i>45] ## using lists comprehension
print(l3)
```
