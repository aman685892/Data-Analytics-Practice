```python
a=("samsung","vivo","redmi","oneplus")
print(a[1:])
```

```python
print(a[:])
```

```python
print(a[2:])
```

```python
print(a[1:4])
```

```python
print(a[::2])
```

```python
print(a[1::2])
```

```python
# to reverse the tuples
print(a[::-1])
```

```python
print(a[2::-1])
```

**Iterations in tuplest**

```python
a=("samsung","vivo","redmi","oneplus")
for i in a:
  print(i)
```

```python
# along with range and length in for loop
for i in range(len(a)):
  print(i)
```

```python
for i in range(len(a)):
  print(a[i])
```

```python
# along with while loop
i=0
while i <len(a):
  print(a[i])
  i+=1
```
