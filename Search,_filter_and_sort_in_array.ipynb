```python
import numpy as np
x=np.array([100,70,80,40])
print(np.sort(x))
```

```python
y=np.array([[100,20,80],[50,10,40]])
print(np.sort(y))
```

```python
## searching in array
z=np.array([10,20,30,40])
print(np.where(z==30))
print(np.where(z%2==0))
```

```python
## searching if the elements are sorted
u=np.array([10,20,30,40])
print(np.searchsorted(u,30))
```

```python
import numpy as np
w=np.array([10,20,30,40])
fa=[True,False,True,False]
k=w[fa]
print(k)
```

```python
L=np.array([12,24,36,48])
fa=L%2==0
b=L[fa]
print(b)
```
