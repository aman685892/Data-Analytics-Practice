```python
import pandas as pd
data=pd.read_excel("/content/ESD.xlsx")
print(data.head(10))
```

```python
gb=data.groupby(["Department","Gender"]).agg({"EEID":"count"})
print(gb)
```

```python
gp1=data.groupby("Country").agg({"Annual Salary":"max"})
print(gp1)
```

```python
gp2=data.groupby(["Country","Gender"]).agg({"Annual Salary":"max","Age":"min"})
print(gp2)
```
