```python
import numpy as np
import statistics as stats
baked_food=[200,300,400,600,100,200]
print(np.mean(baked_food))
print(np.median(baked_food))
print(stats.mode(baked_food))
print(np.std(baked_food))
print(np.var(baked_food))
```

```python
## correlation-coefficient
# -1 represent inverse relationship
# +1 represent positive relationship
# 0 represent no any relationship
Tobacco_consumption=[10,20,30,50,70]
deaths=[100,300,500,700,900]
print(np.corrcoef(Tobacco_consumption,deaths))
```

```python
price=[100,200,50,600,70]
sales=[10,30,50,20,70]
print(np.corrcoef(price,sales))
```
