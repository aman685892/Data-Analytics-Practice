```python
import numpy as np
arr1=np.array([13,23,56])
arr2=np.array([23,78,90])
print(arr1+arr2)
```

```python
x=np.array([[10,20,30],[40,50,60]])
y=np.array([[30,40,50],[10,30,50]])
print(x+y)
print(np.add(x,y))
```

```python
import numpy as np
arr1=np.array([13,23,56])
arr2=np.array([23,78,90])
print(arr1-arr2)
```

```python
import numpy as np
arr1=np.array([13,23,56])
arr2=np.array([23,78,90])
print(arr1*arr2)
print(np.multiply(arr1,arr2))
```

```python
import numpy as np
arr1=np.array([13,23,56])
arr2=np.array([23,78,90])
print(arr1/arr2)
print(np.divide(arr1,arr2))
```

```python
arr1=np.array([13,23,56])
arr2=np.array([2])
print(np.power(arr1,arr2))
```

```python
u=np.array([16,81,9,100])
print(np.sqrt(u))
print(np.sqrt(u).astype(int))
```
