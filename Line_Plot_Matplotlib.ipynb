```python
import matplotlib.pyplot as plt
x=["day-1","day-2","day-3","day-4","day-5","day-6"]
y=[300,400,350,560,679,902]
y1=[890,435,567,853,245,578]
plt.plot(x,y,marker="*",ls="--",color="red",label="week1",alpha=0.2)
plt.plot(x,y1,marker="o",ls="-",color="green",label="week2")
plt.legend()
plt.show()
```
