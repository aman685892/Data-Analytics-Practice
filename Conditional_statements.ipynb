**If-else**

```python
marks=78
if marks>90:
  print("You will get a phone")
else:
  print("no phone for 1 week")

print("Thanku")
```

***if-elif-else statement***

```python
marks=60
if marks>=90:
  print("you will go to a trip:")
elif marks>80 and marks<90:
  print("you will get your phone:")
elif marks>=70 and marks <80:
  print("You will get your books:")
else:
  print("you will not get phone back:")
```

**Nested ift**

```python
marks=89
if marks>90:
  print("you can go to a trip")
  if marks>=98:
    print("you will get a new phone")
else:
  print("No phone")
```
