```python
# string slicing
a="harry potter and the Goblet of fire"
b="0123456789"
print(a)
print(a[0:5])
```

```python
print(a[6:12])
```

```python
print(a[:5]) # by default starts with 0 index
```

```python
print(a[-4:])
```

```python
print(b[::2])
```

```python
print(b[:7:2])
```

```python
# to reverse the string
print(b[::-1])
```

```python
print(b[6::-1])
```
