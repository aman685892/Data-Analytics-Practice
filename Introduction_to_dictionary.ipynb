***Iterations in dictionary***

```python
student={"name":"aman","age":21,"marks":45}
print(student)
```

```python
for i in student:
  print(i) # ie we will get all keys values
```

```python
for i in student:
  print(student[i]) # accessing all values of a given dictionary
```

```python
# using values function to find values of dictionary directly
for x in student.values():
  print(x)
```

```python
# accessing both key values pairs using items() functions
for x,y in student.items():
  print(x,":",y)
```
