```python
import pandas as pd
data1={"Emp Id":["E01","E02","E03","E04","E05","E06"],"Name":["aman","raju","rahul","vijay","prashant","sanju"],"Age":[21,23,34,24,26,28]}
data2={"Emp Id":["E01","E02","E03","E04","E05","E06"],"Salary":[26000,34000,67000,23000,58000,32000]}
df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
print(df1)
print(df2)






```

```python
print(pd.merge(df1,df2,on="Emp Id"))
```

```python
print(pd.merge(df1,df2,on="Emp Id",how="inner"))
```

```python
print(pd.merge(df1,df2,on="Emp Id",how="left"))
```
