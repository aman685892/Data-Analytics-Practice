**Write a program to find sum of all even numbers upto 50**

```python
sum=0
for i in range(1,51):
  if i%2==0:
    sum=sum+i
print(  "Sum of all even numbers upto 50 is:",   sum)
```

**Write a program to write first 20 numbers and their squared numbers**

**Write a program to find sum of first 10 odd numbers using while loop**

```python
sum=0
x=1
while x<20:
  if x%2!=0:
    sum=sum+x
  x+=1
print("Sum of first 100 odd numbers is:",    sum)
```
