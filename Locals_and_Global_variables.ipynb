```python
# Locals Variables
x=24
def hello():
  x=78 # it is not global so its scope is within this block only
  return x
print(hello())
print(x) # its scope only outside
```

```python
x=67
def hello():
  global x
  x=90
  return x
print(hello())
print(x) # here x=90 as it is global so not restricted , its scope is anywhere
```
