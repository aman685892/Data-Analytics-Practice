```python
import matplotlib.pyplot as plt
x=["part-1","part-2","part-3","part-4","part-5","part-6"]
y=[89,98,97,93,95,67]
colors=["red","green","yellow","blue","orange","purple"]
plt.bar(x,y,color=colors ,edgecolor="black"  )
plt.title("Popularity of different Parts of harry potter",fontsize=20)
plt.xlabel("Parts of Harry potter",fontsize=20)
plt.ylabel("Popularity",fontsize=20)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel("/content/expense3 (1).xlsx")
df=pd.DataFrame(data)
print(df)
```

```python
df=df.dropna(subset=["Payment Mode"])

Grouped_by=df.groupby("Payment Mode")["Amount"].sum()
print(Grouped_by)

plt.bar(Grouped_by.index,Grouped_by.values)
plt.show()
```
