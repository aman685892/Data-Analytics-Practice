```python
import numpy as np
x=np.array([1,2,3,4])
print(x)
print(type(x))
```

```python
arr=np.array([[1,2,4,5],[4,6,7,8],[7,9,6,5]]) # we will get 3 rows and 4 columns
print(arr)
```

```python
y=np.array([56,78,34,89])
print(y[0:])
print(y[1:])
print(y[2:4])
print(y[2:])
```

```python
z=np.array([[1,3,4,7],[3,5,6,7]])
print(z[0:3,0:2])
```

```python
print(z[0,1:3]) # 0th index row and columns from 1 to 2
```

```python
u=np.array([[1,3,4,7],[3,5,6,7],[7,9,3,4]])
print(u[2,2:])  # row number with index 2 and columns from 2-3
```

```python
print(np.shape(u))
```

```python
print(np.size(u))
```

```python
print(np.ndim(u))
```
