```python
# concatenate
import numpy as np
x=np.array([2,3,4])
y=np.array([5,6,7])
print(np.concatenate([x,y]))
```

```python
import numpy as np
x=np.array([[2,3,4],[4,5,6]])
y=np.array([[5,6,7],[8,4,3]])
print(np.concatenate([x,y],axis=1))
```

```python
x=np.array([2,3,4])
y=np.array([5,6,7])
print(np.hstack([x,y])) # hstack is for horizontal concatenation
```

```python
print(np.vstack([x,y]))
```

```python
a=np.array([[1,2,3,4,5,6],[2,3,4,5,6,8]])
b=np.array_split(a,3)
print(b)
```
