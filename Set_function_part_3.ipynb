```python
a={"Ironman","Hulk","Thor","Captain America"}
b={"Superman","Batman","Wonder-women"}
c={"Hulk","Thor"}
```

```python
# union
print(a.union(b))
print(a.union(c))
```

```python
# difference
print(a.difference(c)) # remove common elemnets
```

```python
# difference update
a.difference_update(c)
print(a)
```

```python
y=a.intersection(c)
print(y)
```

```python
p={"raju","vinod","shyam"}
q={"raju","shyam"}
r=p.intersection(q)
print(r)
```

```python
p.intersection_update(q)
print(p)
```

```python
u={"a","b","c","d","e"}
v={"b","e"}
k=u.symmetric_difference(v)
print(k)
```
