```python
import numpy as np
x=np.array([1,2,3,4,5,6])
print(np.append(x,9))
```

```python
y=np.array([[10,20,30],[40,50,60]])
print(np.append(y,[100]))
```

```python
print(np.insert(x,1,89))
```

```python
print(np.insert(y,1,[57,37,89]))
```
