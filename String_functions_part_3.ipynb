```python
# endswith() - returns true if the strings ends with the specified value
a="aman kumar"
print(a.endswith('r'))
```

```python
print(a.endswith('k'))
```

```python
# startswith()- method returns true if the string starts with a specified value
name="john raj"
print(name.startswith('j'))
```

```python
print(name.startswith('k'))
```

```python
# swapcase - lowercase becomes uppercase and vice-versa
x="rajkumar is My friendSHWBEJDHE"
print(x.swapcase())
```

```python
# strip() returns trimmed version of the string
a="Harry potter ....    ****"
print(a.strip(".,*"))
```

```python
# split()- splits the string at the specified separator , and returns a list
y="#sjs#hus#mwk#bhwb"
print(y.split('#'))
```

```python
z="hello.my name is aman.i ma . a btech student "
print(z.split('.'))
```

```python
# ljust() returns a left justified version of the string
a="harry potter"
print(a.ljust(20,"%"))
```

```python
# rjust returns a right justified versions of the string
e="harry potter"
h=e.rjust(20,"$")
print(h," is my favourite movie")
```

```python
# replace() returns  a string where a specified value is replaced with a specified value
name1="my name is aman"
print(name1.replace("aman","raj"))
```

```python
# rindex() - searches the string for a specified value and returns the last position of where it was found
a="harry potter and the titanic movie"
print(a.rindex("titanic"))
```

```python
# rfind() searches the string for a specified value and returs the last position of where it was found
a="my name is aman kumar and i will do it"
print(a.rfind("will"))
```
