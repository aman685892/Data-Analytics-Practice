```python
x="My name is aman kumara,I am a btech student"
print(len(x)) # to find the length of string
```

```python
# to find a number of times a character is occuring in string
print(x.count('a'))
```

```python
# to convert each letter into upper case
print(x.upper())
```

```python
# to convert each letter into lower case
print(x.lower())
```

```python
# to find index of any character
print(x.index('b'))
```

```python
# to convert first letter of entire string  to capital
print(x.capitalize())
```

```python
# to convert a string into lower case
print(x.casefold())
```

```python
# to find index of a given character
print(x.find('i'))
```
