```python
# write a python script to print a dictionary where the keys are numbers between 1 and 15  and the values are squares of keys
a={}
for i in range(1,16):
  a[i]=i**2
print(a)
```

```python
# write a program to multiply all the items in a dictionary
a={"a":1,"b":2,"c":3,"d":4,"e":5}
product=1
for i in a:
  product=product*a[i]
print(product)
```
