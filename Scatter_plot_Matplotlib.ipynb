```python
import matplotlib.pyplot as plt
import numpy as np
x=np.random.randint(1,10,50) # it will generate 50 random numbers on x axis between 1 and 9 excluding 10
y=np.random.randint(10,100,50)
color=np.random.randint(10,100,50)
plt.scatter(x,y,marker="*",cmap="viridis",c=color)
plt.colorbar()
plt.show()
```
