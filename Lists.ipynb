**Lists Iteration**

```python
# iteration using for loops
a=["aman","raju","pawan","vinay","rohan"]
for i in a:
  print(i)
```

```python
for i in range(len(a)):
  print(i)
```

```python
for i in range(len(a)):
  print(a[i])
```

```python
# Iteration using while loop
i=0
while i<(len(a)):
  print(a[i])
  i+=1
```

```python
# using short hand for Loops
[print(i) for i in a]
```
