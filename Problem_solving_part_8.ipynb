```python
# write a program to find maximum and minimum values in a set
a={12,34,67,89,54,13}
maximum=max(a)
minimum=min(a)
print("The maximum values in set a: ",maximum)
print("The minimum values in set a: ",minimum)
```

```python
# write a program to find common elements in three lists using sets
a=[1,2,3,4,5]
b=[4,5,6,9]
c=[2,3,4,8,0]
print("The common elements in three lists are :",set(a)&set(b)&set(c))
```

```python
# write  a program to find difference between two sets
a={1,2,3,4,5}
b={2,3,4,8,9}
x=a.difference(b)
print(x)
```

```python
# write a python program to remove  an items from a set if it is present in the set
z={1,2,4,5,6}
z.discard(4)
print(z)
```

```python
# write a python program to check if a set is a subset of another set
a={1,2,3,4,5,6}
b={2,3}
print(b.issubset(a))
```
