```python
# convert the folowing dictionary into JSON fORMAT
import json
Student_data={"name":"aman","age":21,"marks":67}
print(type(Student_data))
data=json.dumps(Student_data)
print(data)
print(type(data))
```

```python
# access the value of age from given JSON data
Student_data=""" {"name":"aman","age":21,"marks":45}"""
print(type(Student_data))
data=json.loads(Student_data) # this function converts json string into dictionary
print(data["age"])
print(type(data))
```

```python
# pretty print following json data
Student_data={"name":"aman","age":21,"marks":45}
data=json.dumps(Student_data,indent=4,separators=(",","="))
print(data)
```
