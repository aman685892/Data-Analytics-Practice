**For Loop**

```python
# for loops
for i in range(1,9):
  print(i)
```

```python
i=4
for i in range(0,8,3):
  print(i)
```

```python
for i in range(1,6):
  print("Hello world")
```

```python
for i in range(1,6,2):
  print(i)
```

```python
# multiplication tables of 7
n=7
for i in range(1,11):
  print(n,"*",i,"=",n*i)
```

```python
num=int(input("Enter the number:"))
for i in range(1,11):
  print(num,"*",i,"=",num*i)
```

**While Loop**

```python
n=1
while n<=5:
  print(n)
  n+=1
```

```python
x=8
while x<=18:
  print(x)
  x+=2
```

```python
# printing tables of 7 using while loop
y=7
i=1
while i<=10:
  print(y,"*",i,"=",y*i)
  i+=1
```

**while true**

```python
while True:
  num1=int(input("Enter the number1:"))
  num2=int(input("Enter the number2:"))
  c=num1+num2
  print(c)
```

```python
while True:
  num1=int(input("Enter the number1:"))
  num2=int(input("Enter the number2:"))
  print(num1+num2)
  c=input("do you want to stop the program:")
  if c=="yes":
    break
```

**Nested Loops**

```python
for i in range(1,4):
  for j in range(1,11):
    print(j,end="")
  print()
```

**For Loops with conditional statements**

```python
for i in range(1,11):
  if i==3:
    print("That song is my favourite:")
  else:
    print(i)
```

```python
## To findout common multiples of 8 and 12 in this range 1-100
for i in range(1,101):
  if i%8==0 and i%12==0:
      print(i)
```

**While Lopps with conditional**

```python
n=1
while n<=10:
  if n==6:
    print("aman")
  else:
    print(n)
  n+=1
```

**Break and continue statements**

```python
for i in range (1,11):
  if i==5:
    continue
  else:
    print(i)
```

```python
for i in range(1,12):
  if i==4:
    break
  else:
    print(i)
print("thanks")
```
