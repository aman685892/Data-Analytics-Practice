```python
a={"Ironman","Hulk","Thor","Captain America"}
b={"Superman","Batman","Wonder-women"}
c={"Hulk","Thor"}
# isdisjoint
print(a.isdisjoint(b))
```

```python
print(a.isdisjoint(c))
```

```python
# issubset
print(a.issubset(b))
```

```python
print(c.issubset(a)) # yes c is a subset of a
```

```python
# issuperset
print(a.issuperset(b))
```

```python
print(a.issuperset(c))
```

```python
# update
a.update(c)
print(a)
```

```python
a.clear()
print(a) # we get an empty set
```
