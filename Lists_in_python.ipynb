```python
my_list=[34,"maan",56.9,'v']
print(my_list)
```

```python
## indexing in list
my_list[1]
```

```python
## slicing in list
my_list[1:3]
```

```python
for x in my_list:
  if x=="maan":
    print(x)
```

```python
l=[12,34,5,67,78]
l.append(89)
print(l)
```

```python
## extend() – add multiple items
print(l)
```

```python
l.extend([23,54])
print(l)
```

```python
print(l)
```

```python
l.insert(1,"aman")
print(l)
```

```python
##remove() – removes first occurrence
l.remove('aman')
l
```

```python
##pop() – removes item by index
l.remove(34) ## since 34 is in list so removed
l
```

```python
##clear() – removes all items
l.clear()
l
```

```python
##del – delete list or element
x=[1,2,3,4,56]
del x[2]
x
```

```python
y=[9,13,90.6,2,-3]
y.sort() ## to sort() all the elements of the list will be of same type
y
```
