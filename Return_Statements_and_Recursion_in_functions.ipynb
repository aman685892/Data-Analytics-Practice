```python
# to find factorial use recursions
def fact(n):
  if n==1:
    return (1)
  else:
    return n*fact(n-1)

print(fact(5))
```

```python
def rec(length,breadth):
  Area=length*breadth
  print(Area)
print(rec(4,6))
```
