**Write  a program to get Fibonacci serirs upto 10 numbers**

```python
a=0
b=1
print(a)
print(b)
for i in range (2,11):
  c=a+b
  a=b
  b=c
  print(c)
```

**Write a program to check if a number is prime or not**

```python
num=int(input("Enter the number:"))
if num<=1:
  print("It is not a prime number")
else:
  for i in range(2,num):
    if num%i==0:
      print("It is not a prime number")
      break
  else:
      print("It is a prime number")
```

**Write a program to find pallindrome of integers**

```python
num=int(input("Enter the number:"))
temp=num
rev=0
while num>0:
  dig=num%10
  rev=rev*10+dig
  num=num//10
if rev== temp:
  print("It is pallindrome:")
else:
  print("It is not pallindrome")
```
