```python
import numpy as np
x=np.array([10,20,30,40])
print(x)
```

```python
print(np.size(x)) # four elemnts are there in the array
print(np.mean(x))
print(np.sum(x))
print(np.min(x))
print(np.max(x))
print(np.cumsum(x))
print(np.cumprod(x))
```

```python
a=[10,20,30,40,50,60]
b=[20,30,40,50,70,90]
price=np.array(a)
quantity=np.array(b)
print(price,quantity)
t=np.cumprod([price,quantity],axis=0)
print(t[1].sum())
```
